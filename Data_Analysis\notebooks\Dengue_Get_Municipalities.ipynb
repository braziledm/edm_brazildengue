## This reads all files, one at a time

```python
years = np.arange(2007,2022,1) # years we need to analyze
id_municipalities = [] # municipalities that have registered dengue over the years ????

for year in years:
  file_path = 'drive/MyDrive/DataBR/Data/DataBR_Processed/dengue_BR_'+str(year)+'.csv'
  file_read = pd.read_csv(file_path,
                        delimiter = ';',
                        index_col=False,
                        parse_dates = ['DT_SIN_PRI','SEM_PRI','DT_NOTIFIC','SEM_NOT'])
  id_municipalities.append(list(file_read['ID_MUNICIP'].unique())) # Array with Municipality IDs for thos municipalities with at least one dengue case reported
```

```python
len(id_municipalities)
```

## Saving the municipalities that had dengue cases to csv

```python
pd.DataFrame(id_municipalities).to_csv('potential_municipalities.csv')
```
