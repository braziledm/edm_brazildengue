## This file translates the codes of selected municipalities to names and states they are from



# Import Packages

```python
import numpy as np
import pandas as pd
```

# Read in data

```python
file_30mun_path = '/content/first_30_municipalities.csv'
file_munbr_path = '/content/municipios_brasileiros_tse.csv'

df_30mun = pd.read_csv(file_30mun_path, sep=",",  encoding='latin1')  #encoding argument allows the code to read in Latin letters/special characters, e.g ã
df_munbr = pd.read_csv(file_munbr_path, sep=",",  encoding='latin1')
```

```python
list_tmp = []
```

# The code below takes as input the municipality code and returns the name of the municipality along with the corresponding state

```python
for codigo in df_munbr['codigo_ibge']:
  list_tmp.append(int(str(codigo)[:-1]))
```

```python
df_munbr['codigo_ibge_sinan'] = list_tmp
```

```python
map_dict = dict(zip(df_munbr['codigo_ibge_sinan'], df_munbr['nome_municipio']))
map_dict_2 = dict(zip(df_munbr['codigo_ibge_sinan'], df_munbr['uf']))
```

```python
df_30mun['Name'] = df_30mun['Municipality'].map(map_dict)
```

```python
df_30mun['State'] = df_30mun['Municipality'].map(map_dict_2)
```
